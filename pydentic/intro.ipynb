```python
from pydantic import BaseModel
from dataclasses import dataclass
```

```python



class classroom(BaseModel):
    students:list[str]
    capacity:int

@dataclass
class classroom2():
    students:list[str]
    capacity:int
```

```python
try:
    classroom1=classroom(students=["Alice","Bob",123],capacity=30)

    #The below one still works sonce pydantic will convert the str to int for capacity
    # classroom1=classroom(students=["Alice","Bob","fly"],capacity="30")
    
    print(classroom1)
 
except Exception as e:
    print("Error:",e)
    print("\n\nPydantic validation worked!\n\n")

#For dataClass validation didn't work
classroom2=classroom2(students=["Alice","Bob",124],capacity=30)
print(classroom2)
```

Pydentic Fields and Validation

```python
from pydantic import Field

class Item(BaseModel):
    name: str=Field(min_length=3, max_length=50)
    price: float=Field(gt=0, lt=1000)
    
item_obj=Item(name="Book",price=25.5)
print(item_obj)
print()
try:
    obj2=Item(name="AB",price=100)  #This will raise validation error
except Exception as e:
    print("Error:",e)
    print("\n\nPydantic field validation worked!\n\n")
```
